```python
bikes.season = bikes.season.astype('str')
bikes.weather_code = bikes.weather_code.astype('str')
```

```python
cols_dict = {
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_c',
    't2':'temp_fells_like_c',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weeked':'is_weeked',
    'season':'season'
}
```

```python
bikes.rename(cols_dict, axis=1, inplace=True)
```

```python
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
bikes
```

```python
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}
```

```python
weather_dict ={
    '1.0':'Clear',
    '2.0':'Scatters Clouds',
    '3.0':'Broken Clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rian with thunderstorm',
    '26.0':'Snowfall',
    '94.0':'Freezing Fog'
}
```

```python
bikes.season = bikes.season.map(season_dict)
bikes.weather = bikes.weather.map(weather_dict)
```

```python
bikes
```

```python
bikes.to_csv('london_bikes.csv')
```
